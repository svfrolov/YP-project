### Выберирем платформы с наибольшими суммарными продажами и построим график по годам.

```python
max_sum = pltf.loc[0:11, 'platform'].values
max_sum
```

```python
color = ['r', 'g', 'b', 'c', 'm', 'y', 'k',  'w', 'orange', 'yellow', 'pink', 'lime']

plt.figure(figsize=(15, 6))
plt.grid()
plt.title(f'Продажи на разных платформах')


m_year = 0
for i in range(0, len(max_sum)):
    platform = df.query('platform == @max_sum[@i]').groupby('year_of_release')['sales'].sum()
    m_year += platform.count()
    platform.plot(x=platform.index, y=platform.values, color=color[i], label=max_sum[i], alpha=0.8, grid=True)

plt.xlabel('Год')
plt.ylabel('Кол-во продаж за год')
plt.legend()
plt.show();
```

### За какой характерный срок появляются новые и исчезают старые платформы?

```python
m_year / len(max_sum)
```

**Получается 11 лет, а из графика видно, что платформы существуют 7-10 лет, если не брать в расчёт минимумы на старте и в конце. DS платформа самая долгая, но пик у неёё как и у других платформ порядка 8 лет. Можно выделить компьютеры, актуальны на всём протяжении выхода игр.**

### Соотношение продаж игр на других платформах

```python
s_scor = {}
for p in top_9['platform']:
    s_scor[p] = df.query('platform==@p')[['sales', 'critic_score', 'user_score']].dropna().reset_index(drop=True)
```

```python
i=0
corr9 = pd.DataFrame(columns=['platform','crit_sale','user_sale'])
for pl_name in s_scor.keys():
    corr9.loc[i, 'platform'] = pl_name
    corr9.loc[i, 'crit_sale'] = s_scor[pl_name]['sales'].corr(s_scor[pl_name]['critic_score'])
    corr9.loc[i, 'user_sale'] = s_scor[pl_name]['sales'].corr(s_scor[pl_name]['user_score'])
    i+=1
```

```python
corr9
```

```python
# ==Зависимости продаж от отзывов Критиков
plt.figure(figsize=(16, 9))
plt.grid()
plt.title('Зависимости продаж от отзывов критиков на разных платформах')
i=0
for p in corr9['platform']:
    xc = s_scor[p]['critic_score'].to_list()
    yc = s_scor[p]['sales'].to_list()
    plt.scatter(xc, yc, label=p, color=color[i], alpha=0.8);
    i+=1 
plt.ylim(0,6)
plt.grid()
plt.xlabel('Отзывы критиков')    
plt.ylabel('Объём продаж')
plt.legend()
plt.show();

# ==Зависимости продаж от отзывов Пользователей
plt.figure(figsize=(16, 9))
plt.grid()
plt.title('Зависимости продаж от отзывов пользователей на разных платформах')
i=0
for p in corr9['platform']:
    xc = s_scor[p]['user_score'].to_list()
    yc = s_scor[p]['sales'].to_list()
    plt.scatter(xc, yc, label=p, color=color[i], alpha=0.8);
    i+=1 
plt.ylim(0,4)
plt.grid()
plt.xlabel('Отзывы пользователей')    
plt.ylabel('Объём продаж')
plt.legend()
plt.show();
```

## Проверяем гипотезы

**Н₀ - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**

**Н₁ - Средние пользовательские рейтинги жанров Action и Sports разные**

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
XOne = df.query(' platform == "XOne" ')['user_score'].dropna().to_list()
PC = df.query(' platform == "PC" ')['user_score'].dropna().to_list()
```

```python
print('Дисперсия XOne =', np.var(XOne), '// Дисперсия PC =', np.var(PC))
```

***Очень точные значения не нужны, укажем 5%**

```python
from scipy import stats as st

alpha = 0.05

results = st.ttest_ind(XOne, PC, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Нулевая гипотеза отвергнута, значить и Н₀ гипотеза отвергнута, следовательно рейтинги этих двух платформ Xbox One и PC различны**

### Средние пользовательские рейтинги жанров Action и Sports разные

```python
Action = df.query(' genre == "Action" ')['user_score'].dropna().to_list()
Sports = df.query(' genre == "Sports" ')['user_score'].dropna().to_list()
```

```python
print('Дисперсия Action =', np.var(Action), '// Дисперсия Sports =', np.var(Sports))
```

**Очень точные значени яне нужны, укажем 5%**

```python
alpha = 0.05

results = st.ttest_ind(Action, Sports, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Нулевая гипотеза отвергнута, значит гипотеза Н₁ отвергнута, средние рейтинги жанров разные**
